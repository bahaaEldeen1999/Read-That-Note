```python
import cv2 as cv
img = cv.imread("imgs/test1.png",0)
template = cv.imread("imgs/template.png",0)
res = cv.matchTemplate(img,template,cv.TM_SQDIFF_NORMED) 
threshold = 0.3
w,h=template.shape
loc = np.where( res >= threshold)
for pt in zip(*loc[::-1]): 
    cv.rectangle(img, pt, (pt[0] + w, pt[1] + h), (0,0,255), 3)
show_images([img])
```
